### Objectives and First and Early Game objective analysis
Let's explore objectives such as towers, kills, dragons, barons, inhibitors. Following this exploration, we will choose to live out rift heralds for our feature engineering and modeling.

For these objectives, we'll just find the number of each per team for now, as well as who which team got the first objective.

```python
def get_first_objective_team(row, objective):
    bObjectives = sorted(row['b' + objective], key=lambda x: x[0])
    rObjectives = sorted(row['r' + objective], key=lambda x: x[0])
    if len(bObjectives) == 0 and len(rObjectives) == 0:
        # print("no objectives on either side, might be invalid game", row)
        # print(row['index'])
        return "nobody"
    elif len(rObjectives) == 0:
        return "blue"
    elif len(bObjectives) == 0:
        return "red"
    elif bObjectives[0][0] <= rObjectives[0][0]:
        return "blue"
    else:
        return "red"
```

```python
games_df['firstTower'] = games_df.apply(
    lambda row: get_first_objective_team(row, "Towers"), axis=1)
games_df['bTowers'] = games_df['bTowers'].map(lambda x: len(x))
games_df['rTowers'] = games_df['rTowers'].map(lambda x: len(x))
print(games_df['firstTower'].value_counts())

groupby = games_df.groupby(['firstTower', 'result']).agg('count')
print(
    "games where blue gets first tower, blue wins {blue_win} games out of ({blue_win}+{red_win}) games: {percent}%"
    .format(blue_win=groupby.loc["blue", "blue"][0],
            red_win=groupby.loc["blue", "red"][0],
            percent=groupby.loc["blue", "blue"][0] /
            (groupby.loc["blue", "blue"][0] + groupby.loc["blue", "red"][0]) *
            100))
print(
    "games where red gets first tower, blue wins {blue_win} games out of ({blue_win}+{red_win}) games: {percent}%"
    .format(blue_win=groupby.loc["red", "blue"][0],
            red_win=groupby.loc["red", "red"][0],
            percent=groupby.loc["red", "blue"][0] /
            (groupby.loc["red", "blue"][0] + groupby.loc["red", "red"][0]) *
            100))
groupby
```

```python
# team = 0 (blue) or 1 (red): games a team won with that objective / total games won by the team
def calc_win_percent(team, objective, df):
    team_total_wins = games_df['result'].value_counts().loc[team]
    df = df[[objective, 'result']]
    win_percent = len(df[(df[objective] == team)
                         & (df['result'] == team)]) / team_total_wins * 100
    return win_percent


objectives_first_win_percent = []
objectives_first = [
    'firstTower', 'firstBlood', 'firstDragon', 'firstBaron', 'firstInhib'
]  # leave out hearald since new
for objective in objectives_first:
    objective_win_percent_blue = calc_win_percent("blue", objective, games_df)
    objectives_first_win_percent.append(objective_win_percent_blue)
    objective_win_percent_red = calc_win_percent("red", objective, games_df)
    objectives_first_win_percent.append(objective_win_percent_red)

for i in range(len(objectives_first)):
    print('%s:' % objectives_first[i])
    print('Team (Blue) - %s win rate ' % objectives_first_win_percent[2 * i])
    print('Team (Red) - %s win rate ' %
          objectives_first_win_percent[2 * i + 1])

team = ["Blue", "Red"]
blue_team = [objectives_first_win_percent[i] for i in range(0, 9, 2)]
red_team = [objectives_first_win_percent[i] for i in range(1, 10, 2)]
win_probability = pd.DataFrame([blue_team, red_team],
                               columns=objectives_first,
                               index=[team]).reset_index()
win_probability = win_probability.rename(columns={'level_0': 'result'})

pos = list(range(len(win_probability['result'])))
width = 0.10

# plotting the bars
fig, ax = plt.subplots(figsize=(10, 5))

# create a bar with firstblood data, in position pos

plt.bar(pos,
        win_probability['firstTower'],
        width,
        label=win_probability.columns[1])

plt.bar([p + width for p in pos],
        win_probability['firstBlood'],
        width,
        label=win_probability.columns[2])

plt.bar([p + width * 2 for p in pos],
        win_probability['firstDragon'],
        width,
        label=win_probability.columns[3])

plt.bar([p + width * 3 for p in pos],
        win_probability['firstBaron'],
        width,
        label=win_probability.columns[4])

plt.bar([p + width * 4 for p in pos],
        win_probability['firstInhib'],
        width,
        label=win_probability.columns[5])

plt.xlim(min(pos) - width, max(pos) + width * 5)
ax.set_ylabel('Win Probability')
ax.set_xlabel('Team')
ax.set_title('Probaility of winning from first objective')
ax.set_xticks([p + 2 * width for p in pos])
ax.set_xticklabels(win_probability['result'])
plt.legend()
```

```python
def count_instances(df, d):
    teams = ["blue", "red"]
    for team in teams:
        for games in df.index:
            for champion in df.loc[games]["team" +
                                          str(0 if team == "blue" else 1)]:
                if str(champion) in d:
                    d[str(champion)]['picked'] += 1
                    if df.loc[games]["result"] == team:
                        d[str(champion)]['won'] += 1
                else:
                    d[str(champion)] = {'picked': 0, 'won': 0}
                    d[str(champion)]['picked'] = 1
                    if df.loc[games]["result"] == team:
                        d[str(champion)]['won'] = 1
    return d


champion_dict = {}
champion_dict = count_instances(total_team, champion_dict)


def champion_win_percent(data):
    for champion in data:
        data[champion]['won'] = (data[champion]['won'] /
                                 data[champion]['picked']) * 100
    return data


champion_dict = champion_win_percent(champion_dict)
sorted_champs_win = sorted(champion_dict.items(),
                           key=lambda x: x[1]['won'],
                           reverse=True)
print(sorted_champs_win[:20])
print("--------")
sorted_champs_picked = sorted(champion_dict.items(),
                              key=lambda x: x[1]['picked'],
                              reverse=True)
print(sorted_champs_picked[:20])
```

```python
# Define a function that retrieves the input's Ranked matches on Summoner's Rift
def retrieve_matches(usernames):
    all_df = pd.DataFrame()
    for username in usernames:
        player = watcher.summoner.by_name(my_region, username)
        print(player)
        player_matches = watcher.match.matchlist_by_account(
            my_region, player['accountId'])
        recent_matches = [
            player_matches["matches"][i]["gameId"]
            for i in range(len(player_matches["matches"]))
            if (player_matches["matches"][i]["queue"] == 420) or (
                player_matches["matches"][i]["queue"] == 440)
        ]  # all matches on Summoner's Rift
        matches_df = pd.DataFrame()

        for i in range(len(recent_matches)):
            time.sleep(1)
            match_detail = watcher.match.by_id(my_region, recent_matches[i])
            participants = []
            for row in match_detail['participants']:
                participants_row = {}
                participants_row['gameId'] = recent_matches[i]
                participants_row['participantId'] = row['participantId']
                participants_row['teamId'] = row['teamId']
                participants_row['champion'] = row['championId']
                participants_row['spell1'] = row['spell1Id']
                participants_row['spell2'] = row['spell2Id']
                participants_row['result'] = row['stats']['win']
                # participants_row['position'] = position_dict[(row['timeline']['lane'],row['timeline']['role'])]
                participants.append(participants_row)

            participants_dict = {
                match_detail['participantIdentities'][i]['participantId']:
                match_detail['participantIdentities'][i]['player']
                ['summonerName']
                for i in range(10)
            }
            summoner_dict = {
                match_detail['participantIdentities'][i]['participantId']:
                match_detail['participantIdentities'][i]['player']
                ['summonerId']
                for i in range(10)
            }

            for row in participants:
                row['championName'] = champ_dict[str(row['champion'])]
                row['spell_1'] = spells_dict[str(row['spell1'])]
                row['spell_2'] = spells_dict[str(row['spell2'])]
                row['summonerName'] = participants_dict[(row['participantId'])]
                row['team'] = team_dict[(row['teamId'])]
                row['summonerId'] = summoner_dict[(row['participantId'])]
                row

            df = pd.DataFrame(participants)
            # df["championTeam"] = df["championName"] + "_" + df["team"]
            df["rank"] = df["summonerId"].apply(lambda x: rank(x))
            df["result"] = df.apply(
                lambda row: 0
                if ((row['result'] == True) and (row['team'] == "Blue")) or (
                    (row['result'] == False) and (row['team'] == "Red")) else
                (1 if (row['result'] == True) and (row['team'] == "Red") or (
                    (row['result'] == False) and
                    (row['team'] == "Blue")) else 2),
                axis=1)
            df["season"] = match_detail["seasonId"]
            df['spell_1'] = df['spell_1'].str.slice(8, )
            df['spell_2'] = df['spell_2'].str.slice(8, )
            df.drop(columns=[
                "participantId", "teamId", "summonerName", "summonerId",
                "spell1", "spell2", "champion", "team"
            ],
                inplace=True)
            matches_df = pd.concat([matches_df, df], ignore_index=True)
        all_df = pd.concat([all_df, matches_df])
    return all_df
```
